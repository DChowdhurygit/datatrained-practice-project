# SEPARATING THE COLUMNS INTO FEATURES AND TARGET

```python
features=df.drop("sales",axis=1)
target=df["sales"]
```

# scaling the data using min-max scaler

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
features=mms.fit_transform(features)
features
```

```python
features.shape
```

THERE ARE 200 ROWS AND 4 COLUMNS

```python
target.shape
```

there are 200 rows and 1 columns in target columns

# APPLYING VARIOUS MODEL

# LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression 
lr=LinearRegression()
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.2,random_state=22)
```

```python
lr.fit(features_train,target_train)
```

```python
pred_test=lr.predict(features_test)
```

```python
print(r2_score(target_test,pred_test))
```

# SUPPORT VECTOR REGRESSOR

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(features_train,target_train)
pred_test=svr.predict(features_test)
print(r2_score(target_test,pred_test))
```

# DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(features_train,target_train)
pred_test=dtr.predict(features_test)
print(r2_score(target_test,pred_test))
```

# KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor()
knr.fit(features_train,target_train)
pred_test=knr.predict(features_test)
print(r2_score(target_test,pred_test))
```

# REGULARIZATION 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.linear_model import Lasso 


parameters={"alpha":[.0001,.001,.01,.1,1,10],"random_state":list(range(0,10))}
ls=Lasso()
clf=GridSearchCV(ls,parameters)
clf.fit(features_train,target_train)

print(clf.best_params_)
```

```python
ls=Lasso(alpha=0.01,random_state=0)
ls.fit(features_train,target_train)
ls.score(features_train,target_train)
pred_ls=ls.predict(features_test)

lss=r2_score(target_test,pred_ls)
lss
```

# Cross validation score

In order to get the best model we need to check the cross validation score

```python
print(cross_val_score(ls,features,target,cv=5).mean())
```

```python
print(cross_val_score(lr,features,target,cv=5).mean())
```

```python
print(cross_val_score(dtr,features,target,cv=5).mean())
```

```python
print(cross_val_score(svr,features,target,cv=5).mean())
```

```python
print(cross_val_score(knr,features,target,cv=5).mean())
```

# ENSEMBLE TECHNIQUE

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

parameters={"criterion":["mse","mae"],"max_features":["auto","sqrt","log2"]}
rf=RandomForestRegressor()
clf=GridSearchCV(rf,parameters)
clf.fit(features_train,target_train)

print(clf.best_params_)
```

```python
rf=RandomForestRegressor(criterion="mae",max_features="auto")

rf.fit(features_train,target_train)
rf.score(features_train,target_train)
pred_rf=rf.predict(features_test)

rfs=r2_score(target_test,pred_rf)
print('R2_score:',rfs*100)


rfc=(cross_val_score(rf,features,target,cv=5).mean())
print("cross_val_score:",rfc*100)
```

# SAVING THE BEST MODEL

```python
import joblib
joblib.dump(rf,"advrf.obj")
```

```python
loaded_model=joblib.load('advrf.obj')
```

```python
result=loaded_model.score(features_test,target_test)
print(result)
```

as we can see the result is giving us 98% accuracy
